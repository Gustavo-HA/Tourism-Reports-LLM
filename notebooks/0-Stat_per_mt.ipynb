```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data = pd.read_csv('../data/raw/reviews.csv')
```

```python
magical_towns = data['Town'].unique().tolist()
```

```python
magical_towns.__len__()
```

```python
def polarity_distribution_per_mt(magical_town):
    df_mt = data[data['Town'] == magical_town]
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_mt, x='Polarity', hue='Type', palette='viridis')
    plt.title(f'Polarity Distribution for {magical_town}. Total Reviews: {df_mt.shape[0]}')
    plt.xlabel('Polarity')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
for mt in magical_towns:
    polarity_distribution_per_mt(mt)
```

```python
types = data['Type'].unique().tolist()
```

```python
# Now also for types
def polarity_distribution_per_type(type_):
    df_type = data[data['Type'] == type_]
    plt.figure(figsize=(10, 6))
    sns.barplot(x=df_type['Polarity'].value_counts().index, 
                y=df_type['Polarity'].value_counts(normalize=True).values,
                legend=False, hue=df_type['Polarity'].value_counts().index,
                palette='viridis')
    plt.title(f'Polarity Distribution for {type_}. Total Reviews: {df_type.shape[0]}')
    plt.xlabel('Polarity')
    plt.ylabel('Frequency')
    plt.yticks(np.arange(0, 1.1, 0.05))
    plt.grid(True)
    plt.show()
```

```python
for type_ in types:
    polarity_distribution_per_type(type_)
```
