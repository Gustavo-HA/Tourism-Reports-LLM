```python
import pandas as pd
```

```python
full_dataset = pd.read_csv("../data/restmex/restmex-corpus.csv", encoding="utf-8")
```

```python
def fix_mojibake(text):
    return text.encode('latin1').decode('utf-8', errors='ignore')

full_dataset['Review'] = full_dataset['Review'].astype(str).apply(fix_mojibake)
full_dataset['Title'] = full_dataset['Title'].astype(str).apply(fix_mojibake)
```

```python
full_dataset.head()
```

```python
full_dataset['text'] = full_dataset['Title'] + " " + full_dataset['Review']
full_dataset = full_dataset[['text', 'Polarity', 'Town', 'Region', 'Type']]
```

```python
full_dataset.to_csv("../data/restmex/restmex-corpus-cleaned.csv", index=False, encoding="utf-8")
```

```python
reduced_dataset = full_dataset.loc[full_dataset['Town'].isin(['Isla_Mujeres', 'Coatepec'])]
```

```python
reduced_dataset.to_csv("../data/restmex/restmex-corpus-reduced.csv", index=False, encoding="utf-8")
```
